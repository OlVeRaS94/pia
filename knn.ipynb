Començarem creant un conjunt aleatori de 10 punts, situats al plà entre el punt (0,0) i (10, 10). Per tal d'obtindre els mateixos punts pseudoaleatoris utilitza una semilla de valor 42.

```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Estableix la semilla per garantir la repetició dels resultats
random.seed(42)

puntos = np.random.uniform( 0, 10, (10, 2))
print(puntos)

punto_referencia = puntos[0]

distancia = np.linalg.norm(puntos - punto_referencia, axis=1)



plt.scatter(puntos[:, 0], puntos[:, 1], color='blue', label='Puntos')
plt.scatter(punto_referencia[0], punto_referencia[1])


plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
